Vamos a hacer una flor rosada con Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt

# Configuración de los datos
# theta representa el ángulo de 0 a 2π
theta = np.linspace(0, 2 * np.pi, 1000)

# Ecuación de la rosa: r = cos(k * theta)
# Si 'k' es entero, determina el número de pétalos
k = 5
r = np.cos(k * theta)

# Crear la figura con proyección polar
plt.figure(figsize=(6, 6))
ax = plt.subplot(111, projection='polar')

# Dibujar la flor
ax.plot(theta, r, color='deeppink', linewidth=2, label=f'Flor de {k} pétalos')

# Rellenar los pétalos para que se vea mejor
ax.fill(theta, r, color='hotpink', alpha=0.3)

# Ajustes estéticos
ax.set_yticklabels([]) # Quita los números de los radios
ax.set_xticklabels([]) # Quita los ángulos
ax.spines['polar'].set_visible(False) # Quita el borde circular
ax.set_title("Flor Matemática en Matplotlib", va='bottom', fontsize=15)

plt.show()
```

#La grafica de onda de sonido

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la onda
frecuencia = 5  # Hercios (ciclos por segundo para la visualización)
amplitud = 1    # Volumen
duracion = 1    # Segundos
frecuencia_muestreo = 500 # Puntos por segundo

# Crear el eje del tiempo
t = np.linspace(0, duracion, frecuencia_muestreo)

# Generar la onda de sonido (función seno)
y = amplitud * np.sin(2 * np.pi * frecuencia * t)

# Configuración de la gráfica
plt.figure(figsize=(10, 4))
plt.plot(t, y, color='blue', linewidth=2)

# Estética de la gráfica
plt.title(f"Onda de Sonido Pura (Seno) - {frecuencia} Hz", fontsize=14)
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud")
plt.grid(True, alpha=0.3)
plt.axhline(0, color='black', linewidth=1) # Línea central

plt.show()

# Dos frecuencias distintas
f1, f2 = 5, 12
y_combinada = np.sin(2 * np.pi * f1 * t) + 0.5 * np.sin(2 * np.pi * f2 * t)

plt.figure(figsize=(10, 4))
plt.plot(t, y_combinada, color='purple')
plt.title("Interferencia de dos ondas de sonido")
plt.show()
```

Esta onda esta genial

```python
import numpy as np
import matplotlib.pyplot as plt

def espiral_fibonacci_perfecta(n_puntos=1000):
    # El número áureo (phi)
    phi = (1 + 5**0.5) / 2

    # Creamos el ángulo (theta) que va girando
    # Aumentar el rango de theta para más vueltas
    theta = np.linspace(0, 5 * np.pi, n_puntos)

    # Ecuación de la espiral dorada: r = a * e^(b * theta)
    # donde b depende de phi
    b = np.log(phi) / (np.pi / 2)
    r = np.exp(b * theta)

    # Convertir coordenadas polares a cartesianas para graficar
    x = r * np.cos(theta)
    y = r * np.sin(theta)

    # Graficar
    plt.figure(figsize=(8, 8))
    plt.plot(x, y, color='firebrick', linewidth=2, label="Espiral Dorada")

    # Estética
    plt.axhline(0, color='black', lw=0.5)
    plt.axvline(0, color='black', lw=0.5)
    plt.title("El 'Caracol' de Fibonacci (Espiral Dorada)", fontsize=14)
    plt.axis('equal')
    plt.grid(True, linestyle=':', alpha=0.6)
    plt.show()

espiral_fibonacci_perfecta()
```
